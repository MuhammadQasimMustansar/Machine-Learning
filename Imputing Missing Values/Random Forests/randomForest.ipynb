# Random Forests for Imputing Missing Values

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
```

```python
df = sns.load_dataset("titanic")
df.head() 
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.drop('deck', axis=1, inplace=True)

df.isnull().sum().sort_values(ascending=False)
```

```python
imputer = SimpleImputer(strategy="most_frequent")
df[["embarked", "embark_town"]] = imputer.fit_transform(df[["embarked", "embark_town"]])

df.isnull().sum().sort_values(ascending=False)
```

```python
column_to_encode = ['sex', 'embarked', 'who', 'class', 'embark_town', 'alive']

Label_Encoder = {}

for col in column_to_encode:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

    Label_Encoder[col] = le

df.head()
```

```python
# spliting data into two parts
with_missing = df[df["age"].isna()]
without_missing = df.dropna()
```

```python
with_missing
```

```python
without_missing
```

```python
X = without_missing.drop('age', axis=1)
y = without_missing["age"]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

rfr = RandomForestRegressor(n_estimators=100, random_state=42)
rfr.fit(X_train, y_train)

rfr.predict(X_test)
```

```python
y_pred = rfr.predict(with_missing.drop('age', axis=1))
with_missing["age"] = y_pred
with_missing.head()
```

```python
complete_data = pd.concat([without_missing, with_missing], axis=0)
print("shape of complete data frame:", complete_data.shape)
complete_data.head()
```

```python
for col in column_to_encode:
    le = Label_Encoder[col]
    complete_data[col] = le.inverse_transform(df[col])

complete_data.head()    
```

```python
complete_data.isnull().sum().sort_values(ascending=False)
```
