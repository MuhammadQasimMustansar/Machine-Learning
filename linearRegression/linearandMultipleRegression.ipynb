# Regression Matrices

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np 
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

df = sns.load_dataset("tips")
df.head()

X = df[['total_bill']]
y = df['tip']

plt.title("Linear Regression", color="green")
plt.xlabel("total_bill", color="red")
plt.ylabel("tip", color="red")

# plt.scatter(X, y)
# plt.plot(df.total_bill, lr.predict(df[['total_bill']]), color="blue")

# plt.show()

sns.regplot(x=X, y=y, color="blue", line_kws={"color": "red"})

lr = LinearRegression()

lr.fit(df[['total_bill']], df.tip)

pred = lr.predict([[5000]])

# Make Mannual Calculation

print(f"predict value is: {pred}")
print(f"coefficient value is: {lr.coef_}")
print(f"intercept value is: {lr.intercept_}")

lr.coef_ * 5000 + lr.intercept_
```
